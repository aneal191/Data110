#3. Visualizing Theft Offenses by Day of the Week

Task: Which day of the week has the highest number of theft arrests? What might be the reasons for this trend?

```python
# Load the dataset
file_path = 'https://raw.githubusercontent.com/aneal191/Data110/refs/heads/main/Dataset/Daily_Arrests_20241001%20(1).csv'
data = pd.read_csv(file_path)

# Convert 'ARREST DATE' to datetime format for accurate manipulation
data['ARREST DATE'] = pd.to_datetime(data['ARREST DATE'], errors='coerce')

# Filter the dataset for theft-related offenses
theft_data = data[data['OFFENSE'].str.contains("THEFT", case=False, na=False)]

# Extract the day of the week from the 'ARREST DATE' column
theft_data.loc[:, 'Day of Week'] = theft_data['ARREST DATE'].dt.day_name()

# Count the number of theft incidents by day of the week
theft_by_day = theft_data['Day of Week'].value_counts()

# Visualize the theft incidents by day of the week
plt.figure(figsize=(10, 6))
theft_by_day.sort_index().plot(kind='bar', rot=45)
plt.title('Theft Arrests by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Arrests')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Find the day with the highest number of arrests
most_common_day = theft_by_day.idxmax()
most_common_day_count = theft_by_day.max()

# Output the results
print(f"The day with the highest number of theft arrests is {most_common_day} with {most_common_day_count} arrests.")
```

Results:
Day with the Highest Number of Theft Arrests: Monday

Number of Arrests on that Day: 21

This trend might reflect increased activity or reporting of theft incidents following weekends, when people may notice missing items or report incidents that occurred during leisure time. Let me know if you need further insights! ​

#4. Data Analysis:

Conducted exploratory data analysis (EDA) to understand trends and patterns.
Created visualizations to explore arrest trends and demographics.

```python
# Load the dataset
file_path = 'https://raw.githubusercontent.com/aneal191/Data110/refs/heads/main/Dataset/Daily_Arrests_20241001%20(1).csv'
data = pd.read_csv(file_path)

# Convert 'ARREST DATE' to datetime format for accurate manipulation
data['ARREST DATE'] = pd.to_datetime(data['ARREST DATE'], errors='coerce')

# Filter the dataset for theft-related offenses
theft_data = data[data['OFFENSE'].str.contains("THEFT", case=False, na=False)]

# --- Analysis 2: Demographic Insights (Age Range) ---

# Find the range of ages involved in theft arrests
age_range = (theft_data['AGE'].min(), theft_data['AGE'].max())
print(f"Age range for individuals involved in theft: {age_range[0]} to {age_range[1]} years old")

# Visualization: Age distribution of individuals involved in theft
plt.figure(figsize=(10, 6))
theft_data['AGE'].hist(bins=15, edgecolor='black')
plt.title('Age Distribution of Individuals Involved in Theft')
plt.xlabel('Age')
plt.ylabel('Number of Individuals')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
