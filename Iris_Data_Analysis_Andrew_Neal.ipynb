# **1. Data Visualization**

#1.1 Pair Plot
Create a pair plot of all features in the Iris dataset, color-coding by species. Include an overall title for the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
iris_df = pd.read_csv("https://raw.githubusercontent.com/aneal191/Data110/refs/heads/main/iris.csv")

# Create a pair plot with Seaborn
sns.pairplot(iris_df, hue="species")

# Add an overall title
plt.suptitle("Pair Plot of Iris Dataset Features by Species", y=1.02)

# Display the plot
plt.show()
```


#1.2 Bar Plot

Create a bar plot showing the average sepal length for each iris species.




```python
# Calculate the average sepal length for each species
average_sepal_length = iris_df.groupby("species")["sepal_length"].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=average_sepal_length, x="species", y="sepal_length")

# Add title and labels
plt.title("Average Sepal Length for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")

# Display the plot
plt.show()
```

# 1.3 Box Plot

Create a box plot showing the distribution of petal widths for each species.

```python
# Create a box plot showing the distribution of petal widths for each species
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris_df, x="species", y="petal_width")

# Add title and labels
plt.title("Distribution of Petal Widths for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Petal Width")

# Display the plot
plt.show()
```

# 1.4 Grouped Bar Plot

Create a grouped bar plot showing the average sepal and petal lengths by species.

```python
# Calculate the average sepal and petal lengths for each species
average_lengths = iris_df.groupby("species")[["sepal_length", "petal_length"]].mean().reset_index()

# Melt the DataFrame to make it suitable for a grouped bar plot
average_lengths_melted = average_lengths.melt(id_vars="species",
                                              value_vars=["sepal_length", "petal_length"],
                                              var_name="Measurement",
                                              value_name="Average Length")

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=average_lengths_melted, x="species", y="Average Length", hue="Measurement")

# Add title and labels
plt.title("Average Sepal and Petal Lengths by Species")
plt.xlabel("Species")
plt.ylabel("Average Length")

# Display the plot
plt.show()
```

# 1.5 Violin Plot

Create a violin plot showing the distribution of sepal length for each species.

```python
# Create a violin plot showing the distribution of sepal length for each species
plt.figure(figsize=(8, 6))
sns.violinplot(data=iris_df, x="species", y="sepal_length")

# Add title and labels
plt.title("Distribution of Sepal Length for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")

# Display the plot
plt.show()
```

# **2. Data Analysis**

# 2.1 Aggregation

a) Calculate and display the mean, minimum, and maximum values of all numerical features (sepal length, sepal width, petal length, petal width) for each species.

```python
# Calculate the mean, minimum, and maximum values for each species and each numerical feature
summary_stats = iris_df.groupby("species").agg({
    "sepal_length": ["mean", "min", "max"],
    "sepal_width": ["mean", "min", "max"],
    "petal_length": ["mean", "min", "max"],
    "petal_width": ["mean", "min", "max"]
})

# Display the results
print(summary_stats)
```

b) Count and display the number of samples in each species.

```python
# Count the number of samples in each species
species_counts = iris_df["species"].value_counts()

# Display the results
print(species_counts)
```
