# Part 2: Fare and Survival Analysis
Task 2: Fare Distribution

Plot a histogram and KDE for the distribution of fares paid by passengers.

```python
# Plotting a histogram for the 'fare' column
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], kde=True, bins=30, color='skyblue')

# Adding labels and title
plt.title('Distribution of Fares Paid by Titanic Passengers', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting a KDE for the 'fare' column
sns.kdeplot(titanic['fare'], fill=True, color='skyblue')
```

Question: What trends or outliers do you observe in the fare distribution? How might fare impact survival rates?

**Observations on Fare Distribution:**
**Most passengers paid low fares, with a few outliers who paid significantly higher amounts. The fare distribution is right-skewed, reflecting the higher fares paid by first-class passengers.**

**Some passengers paid unusually high fares, likely indicating premium first-class accommodations.**

**Impact on Survival:**
**Higher Fares = Higher Survival: First-class passengers, who paid more, had better access to lifeboats and higher survival rates.**

**Lower Fares = Lower Survival: Third-class passengers, who paid lower fares, faced more barriers to survival due to limited access and social hierarchies.**

**Fare correlates with class and survival, with higher fares linked to better outcomes.**

Task 3: Fare and Survival Rates

Create a KDE plot to compare the fare distributions for survivors and non-survivors.

```python
# Plotting a KDE to compare fare distributions for survivors and non-survivors
plt.figure(figsize=(10, 6))

# KDE for survivors
sns.kdeplot(titanic[titanic['survived'] == 1]['fare'], fill=True, label='Survived', color='green')

# KDE for non-survivors
sns.kdeplot(titanic[titanic['survived'] == 0]['fare'], fill=True, label='Did Not Survive', color='red')

# Adding labels and title
plt.title('Fare Distribution for Survivors vs Non-Survivors', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()







```

Question: Does fare seem to have influenced survival? What do the distributions suggest about socioeconomic status and survival?

**Yes, fare seems to have influenced survival based on the distributions. The KDE plot for survivors likely shows a peak at higher fares, while non-survivors tend to have lower fares. This suggests that passengers who paid more (typically first-class) had better chances of survival.**

**Socioeconomic Status and Survival:**
**The distributions highlight a strong correlation between socioeconomic status and survival:**

**Higher fares (indicating first-class passengers) correlate with a higher survival rate, likely due to better access to lifeboats and priority during evacuation.**
**Lower fares (typically third-class passengers) are associated with a lower survival rate, possibly due to restricted access to lifeboats and being in lower decks.**
**This reflects how socioeconomic status played a significant role in determining survival during the Titanic disaster.**

# Part 3: Class, Fare, and Survival Insights

```python
# Creating histograms and KDE plots for fare distributions across different passenger classes (Pclass)
plt.figure(figsize=(12, 8))

# Histogram and KDE for first-class passengers
plt.subplot(2, 1, 1)
sns.histplot(titanic[titanic['pclass'] == 1]['fare'], kde=True, color='blue', bins=30)
plt.title('Fare Distribution - First Class', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Histogram and KDE for second-class passengers
plt.subplot(2, 1, 2)
sns.histplot(titanic[titanic['pclass'] == 2]['fare'], kde=True, color='green', bins=30)
plt.title('Fare Distribution - Second Class', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.tight_layout()
plt.show()

# Creating KDE plots to compare fare distributions across different passenger classes (Pclass)
plt.figure(figsize=(12, 8))

# KDE for First Class
sns.kdeplot(titanic[titanic['pclass'] == 1]['fare'], fill=True, label='First Class', color='blue')

# KDE for Second Class
sns.kdeplot(titanic[titanic['pclass'] == 2]['fare'], fill=True, label='Second Class', color='green')

# KDE for Third Class
sns.kdeplot(titanic[titanic['pclass'] == 3]['fare'], fill=True, label='Third Class', color='red')

# Adding labels and title
plt.title('KDE of Fare Distribution Across Passenger Classes', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```

Question: What patterns do you observe in fare distribution and survival rates across different classes? How does this reflect the socioeconomic divisions on the Titanic?

**First Class: Paid the highest fares and had the highest survival rate due to better access to lifeboats.**

**Second Class: Paid moderate fares and had a middle-range survival rate.**

**Third Class: Paid the lowest fares and had the lowest survival rate, with limited access to lifeboats.**

**Socioeconomic Divisions:**
**Higher fare passengers had better access to safety, reflecting the stark socioeconomic divisions on the Titanic, where wealthier passengers had greater chances of survival.**

# Part 4: Age and Survival
Task 5: Age of Survivors and Non-Survivors

Create KDE plots to compare the age distributions of survivors and non-survivors.



```python
# Plotting a KDE to compare age distributions for survivors and non-survivors
plt.figure(figsize=(10, 6))

# KDE for survivors
sns.kdeplot(titanic[titanic['survived'] == 1]['age'], fill=True, label='Survived', color='green')

# KDE for non-survivors
sns.kdeplot(titanic[titanic['survived'] == 0]['age'], fill=True, label='Did Not Survive', color='red')

# Adding labels and title
plt.title('Age Distribution for Survivors vs Non-Survivors', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

Question: What trends do you notice in the age distribution of survivors compared to those who did not survive? How might age have impacted survival rates?


**The KDE plot for survivors shows a higher concentration of younger passengers and children, reflecting the "women and children first" policy during the evacuation. Whereas the age distribution for non-survivors is more evenly spread, with a greater proportion of adults and elderly passengers not surviving.**

**Children and younger passengers had higher survival rates, likely due to being prioritized during the lifeboat evacuation.**

**Older passengers were less likely to survive, possibly due to lower prioritization and physical limitations that hindered their escape.**

**Age appears to have played a significant role, with younger passengers having a better chance of survival.**

Task 6: Pyramid Histogram for Age Survival by Sex

Create a pyramid-style histogram (also known as a population pyramid) for age, showing survival data split by sex. If possible, attempt to find or create one.

```python
# Subsetting the data for female survivors and non-survivors
female_survived = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)]
female_did_not_survive = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 0)]

# Create a pyramid-style histogram
plt.figure(figsize=(12, 8))

# Plot for females who survived and did not survive
plt.hist(female_survived['age'], bins=20, orientation='horizontal', color='blue', alpha=0.6, label='Female Survived')
plt.hist(female_did_not_survive['age'], bins=20, orientation='horizontal', color='orange', alpha=0.6, label='Female Did Not Survive')

# Adding labels and title
plt.title('Pyramid-Style Age Distribution by Sex and Survival', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Subsetting the data for male survivors and non-survivors
male_survived = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 1)]
male_did_not_survive = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 0)]

# Create a pyramid-style histogram
plt.figure(figsize=(12, 8))

# Plot for males who survived and did not survive
plt.hist(male_survived['age'], bins=20, orientation='horizontal', color='green', alpha=0.6, label='Male Survived')
plt.hist(male_did_not_survive['age'], bins=20, orientation='horizontal', color='red', alpha=0.6, label='Male Did Not Survive')

# Adding labels and title
plt.title('Pyramid-Style Age Distribution by Sex and Survival', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

Question: What can you infer from the age and sex distribution in relation to survival rates?

**Women and Children First: Females and younger passengers had higher survival rates, reflecting the "women and children first" policy.**

**Male Survival: Adult males had much lower survival rates, as they were less prioritized during evacuation.**

**Age Impact: Younger passengers had a better chance of survival, while older passengers, especially males, had lower survival rates.**

**Overall, gender and age strongly influenced survival, with females and children being more likely to survive.**

Task 7 (Optional): Bar Graph, Line Plot, and Scatter Plot
Create a bar graph, line plot, and scatter plot using the Titanic dataset.
Suggestions: You can explore variables like survival rates, passenger class, age, or fare in these plots.

```python
# Bar graph: Survival rate by passenger class
plt.figure(figsize=(12, 8))
sns.barplot(x='pclass', y='survived', data=titanic, ci=None)
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```

```python
# Scatter plot: Fare vs. Age, colored by survival
plt.figure(figsize=(12, 8))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic, palette='coolwarm')
plt.title('Fare vs Age Colored by Survival', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.show()
```

What additional insights do these visualizations provide that weren't evident from the histograms and KDE plots?


**The bar graph clearly shows survival rates across passenger classes, highlighting the higher survival rate for first-class passengers. The line plot reveals the relationship between age and fare, showing how fare changes across ages in different classes—something not seen in KDE or histograms. The scatter plot combines fare, age, and survival, making it easier to see how these variables interact (e.g., younger, wealthier passengers having better survival odds).**
