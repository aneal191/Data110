```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#load the dataset

#df=pd.read_csv('Dataset.csv')
df=pd.read_csv('https://raw.githubusercontent.com/aneal191/Data110/d4281ea33e81593c07431ec3885597546bac839a/Dataset/Dataset.csv')
print(df.head())
```

```python
# Extract the data for plotting
offenses = df['Offense']
rates_2023 = df['Offenses per 100,000 People (2023)']

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(offenses, rates_2023, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Crime Rates per 100,000 People in 2023', fontsize=16)
plt.xlabel('Type of Offense', fontsize=12)
plt.ylabel('Offenses per 100,000 People', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# Extract the data for plotting
offenses = df['Offense']
changes_in_rates = df['Change in Offenses per 100,000 People (2022-2023)']

# Create a horizontal bar chart for the change in offense rates
plt.figure(figsize=(10, 6))
bars = plt.barh(offenses, changes_in_rates, color=['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#A1FF33', '#33A1FF', '#FF5733'], edgecolor='black')

# Customize the plot with labels and title
plt.title('Change in Offense Rates (2022-2023)', fontsize=16)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=12)
plt.ylabel('Type of Offense', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

Why does this bar graph make more sense than Task 1's bar graph?

The horizontal bar graph showing the change in offense rates is more insightful than the bar graph of total crime rates because it highlights trends over time.

Absolute numbers can be misleading: A high crime rate (e.g., larceny) might still be improving, while a low rate (e.g., murder) could be worsening. Total numbers don't reflect whether crime is getting better or worse.

Percentage change shows how crime trends are evolving, making it easier to prioritize issues. An increase (e.g., motor vehicle theft) signals growing concern, while a decrease (e.g., larceny) indicates improvement.

In short, change over time provides deeper insights and helps assess policy effectiveness, while absolute numbers offer only a snapshot.








Improve the clarity of your bar graph:

Create multiple versions of the bar chart for Task 2, each time making the chart more clear and readable.
Experiment with different color schemes, labels, and gridlines.
In each iteration, focus on improving the ease with which someone can quickly understand the trends in crime rates.


```python
# Data for offenses and change in offense rates (2022-2023)
offenses = df['Offense']
changes_in_rates = df['Change in Offenses per 100,000 People (2022-2023)']

# First iteration: Enhance readability
plt.figure(figsize=(10, 6))
bars = plt.barh(offenses, changes_in_rates, color='lightblue', edgecolor='black')

# Customize the plot
plt.title('Change in Offense Rates (2022-2023)', fontsize=18)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Offense', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='-', alpha=0.8)

# Display the first improved chart
plt.tight_layout()
plt.show()
```

```python
# Second iteration: Use contrasting colors for each bar
plt.figure(figsize=(10, 6))
bars = plt.barh(offenses, changes_in_rates, color=['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#A1FF33', '#33A1FF', '#FFB533'], edgecolor='black')

# Add data labels to each bar
for i, v in enumerate(changes_in_rates):
    plt.text(v, i, f'{v}%', color='black', va='center', fontsize=12)

# Customize the plot
plt.title('Change in Offense Rates (2022-2023)', fontsize=18)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Offense', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the second improved chart
plt.tight_layout()
plt.show()
```

```python
# Third iteration: Simplify the color scheme and use a subtle grid
plt.figure(figsize=(10, 6))
bars = plt.barh(offenses, changes_in_rates, color='cornflowerblue', edgecolor='black')

# Add data labels to each bar
for i, v in enumerate(changes_in_rates):
    plt.text(v, i, f'{v}%', color='black', va='center', fontsize=12)

# Customize the plot
plt.title('Change in Offense Rates (2022-2023)', fontsize=18)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Offense', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle=':', alpha=0.5)

# Display the third improved chart
plt.tight_layout()
plt.show()
```
