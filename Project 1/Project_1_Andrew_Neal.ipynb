# Introduction to the Dataset
The "Drug Deaths: A Story of Growing Opioid Abuse" dataset contains records of drug-related deaths, including demographics, substances involved, and locations. It provides insights into the opioid crisis, with details on age, sex, race, and substances found in the deceased.

# Why This Dataset Was Chosen
This dataset was selected to explore the growing opioid epidemic and its impact. It offers valuable information to analyze trends in overdose deaths, helping to better understand the crisis and support prevention efforts.

```python
import matplotlib.pyplot as plt
import pandas as pd

#load the dataset

data = pd.read_csv('https://raw.githubusercontent.com/aneal191/Data110/refs/heads/main/drug-deaths-a-story-of-growing-opiod-abuse_kaggle.csv')
print(data.head())
```

```python
# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Create a scatter plot of Age vs Date
plt.figure(figsize=(10,6))
plt.scatter(data['Date'], data['Age'], alpha=0.5)
plt.title('Scatter Plot of Age vs Date of Death')
plt.xlabel('Date of Death')
plt.ylabel('Age of Deceased')
plt.grid(True)
plt.show()
```


The scatter plot illustrates the relationship between the age of deceased individuals and the date of their death. The distribution of points shows a broad range of ages across time, with no specific age group or period standing out significantly. However, the relatively even spread suggests that opioid-related deaths affect individuals across different age groups and years, highlighting the widespread and ongoing nature of the opioid crisis.

```python
# Group the data by Date to get the number of deaths on each date
death_counts_by_date = data.groupby('Date').size()

# Plot the line chart
plt.figure(figsize=(10,6))
plt.plot(death_counts_by_date.index, death_counts_by_date.values, marker='o', linestyle='-', color='b')
plt.title('Number of Deaths Over Time')
plt.xlabel('Date of Death')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The line chart shows the trend in drug-related deaths over time. Peaks in the chart may indicate specific periods where drug-related fatalities increased, potentially due to the rise in opioid abuse or other contributing factors. The overall trend can help identify time periods where interventions might have been most needed or effective. This visualization highlights the continuing urgency of addressing the opioid crisis.








```python
# Creating a histogram to show the distribution of ages of the deceased
plt.figure(figsize=(10,6))
plt.hist(data['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages of Deceased Individuals')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```


The histogram illustrates the distribution of ages among individuals who died due to drug-related causes. It shows which age groups are most frequently affected, with potential peaks indicating particularly vulnerable age ranges. This insight helps identify the age demographics most at risk from the opioid crisis, informing targeted prevention and intervention strategies.
