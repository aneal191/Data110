#Final Project

####This project is an exploratory data analysis of AirBNB in the 5 burroughs of New York in year 2019.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Airbnb dataset
data = pd.read_csv('https://raw.githubusercontent.com/aneal191/Data110/refs/heads/main/Dataset/airbnb_ny19.csv')

# Remove extreme outliers for price visualization clarity
filtered_data = data[data['price'] < 1000]

# Bar Chart of Listings by Borough
plt.figure(figsize=(8, 6))
borough_counts = data['neighbourhood_group'].value_counts()
borough_counts.plot(kind='bar', color='green', edgecolor='none')
plt.title('Number of Listings by Borough', fontsize=16)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.box(False)
plt.tight_layout()
plt.show()

```

####This bar chart is cool because it's clean-looking and clear. There are no borders or gridlines. It looks good. We can clearly see Manhattan has the most number of listings and Staten Island has the fewest.

```python
# Pie Chart of Room Types
plt.figure(figsize=(8, 6))
room_type_counts = data['room_type'].value_counts()
room_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('dark'), textprops={'fontsize': 10})
plt.title('Distribution of Room Types', fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.show()

```

####This pie chart is cool. Look at those dark colors. Also, it's surprising more entire homes get rented out than private rooms. I guess people aren't into sharing rooms. But isn't it a nice pie chart?

```python
# Boxplot of Prices by Borough
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='neighbourhood_group', y='price', hue='neighbourhood_group',
            palette='Set1', dodge=False, legend=False)
plt.title('Price Distribution by Borough', fontsize=16)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 1000)
sns.despine(left=True, bottom=True)
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()

```

####Look at those colors. Solid.
####Surprise, surprise Manhattan has the highest price. As for the Bronx, no one wants to visit that place, it's a sh*hole. People get stabbed there.


```python
# Scatter Plot of Availability vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x='availability_365', y='price', hue='neighbourhood_group', alpha=0.6, s=40)
plt.title('Availability vs. Price by Borough', fontsize=16)
plt.xlabel('Availability (Days per Year)', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend(title='Borough', fontsize=10, title_fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

####This chart is bananas. B-a-n-a-n-a-s. So much information. It really doesn't tell us anything. Well...airbnbs are available any time of year. I guess that's surprising.

```python
# Histogram of Prices
plt.figure(figsize=(10, 6))

# Remove extreme outliers for better visualization
filtered_prices = data[data['price'] < 1000]

# Histogram
sns.histplot(data=filtered_prices, x='price', bins=30, kde=True, color='brown', edgecolor='none')
plt.title('Distribution of Listing Prices', fontsize=16)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

####And last but not least, an amazing histogram! Look at that. Have you ever seen a histogram like this? It's magical. It shows us the number of listings increases as price decreases. That means there are fewer listings at the high price range. Maybe rich people don't care about making money on airbnb because they're too busy buying oil or floating on their yachts on the coast of Bali. Well, ta-da! A histogram.
