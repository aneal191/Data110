```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# style
print(plt.style.available)
plt.style.use('seaborn-whitegrid')
```

```python
# 2. Basic Line Plot
# Define the quadratic function
def f(x):
    return x**2 - 2*x + 1

# Define the range for x values
x_values = np.linspace(-2, 4, 100)
y_values = f(x_values)

# Create the plot
plt.plot(x_values, y_values)
plt.title("Plot of f(x) = x^2 - 2x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
# 3. Multiple Lines

# Define the functions
def f(x):
    return x**3

def g(x):
    return np.exp(x)

def h(x):
    return np.log(x)

# Define the range for x values
x_values = np.linspace(0.1, 4, 100)  # start from 0.1 to avoid log(0)

# Calculate y values for each function
f_values = f(x_values)
g_values = g(x_values)
h_values = h(x_values)

# Create the plot
plt.plot(x_values, f_values, label='f(x) = x^3', color='blue')
plt.plot(x_values, g_values, label='g(x) = e^x', color='green')
plt.plot(x_values, h_values, label='h(x) = log(x)', color='red')

# Add a title, labels, and a legend
plt.title("Plot of f(x) = x^3, g(x) = e^x, and h(x) = log(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
# 4. Styling Lines

# Define the functions
def f(x):
    return x**2

def g(x):
    return x**3

def h(x):
    return x**4

def i(x):
    return x**5

# Define the range for x values
x_values = np.linspace(-2, 2, 100)

# Create the plot with different color specifications and line styles
plt.plot(x_values, f(x_values), label='f(x) = x^2', color='blue', linestyle='-')  # Named color
plt.plot(x_values, g(x_values), label='g(x) = x^3', color='#FF5733', linestyle='--')  # Hex color code
plt.plot(x_values, h(x_values), label='h(x) = x^4', color=(0.2, 0.8, 0.2), linestyle='-.')  # RGB tuple
plt.plot(x_values, i(x_values), label='i(x) = x^5', color='purple', linestyle=':')  # Named color

# Add a title, labels, and a legend
plt.title("Plot of f(x), g(x), h(x), and i(x) with different styles")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
# 5. Axes Limits and Scaling

# Define the function
def f(x):
    return 1 / (x**2 + 1)

# Define the range for x values
x_values = np.linspace(-10, 10, 400)

# Calculate y values for the function
y_values = f(x_values)

# First plot with custom x and y limits
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x) = 1 / (x^2 + 1)', color='blue')
plt.xlim(-10, 10)
plt.ylim(0, 1.1)  # Y limit slightly above 1 to showcase the function
plt.title("Plot of f(x) = 1 / (x^2 + 1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# Second plot with reversed x and y axes
plt.figure(figsize=(8, 6))
plt.plot(y_values, x_values, label='Reversed Axes: f(x) = 1 / (x^2 + 1)', color='green')
plt.title("Plot of f(x) with Reversed Axes")
plt.xlabel("f(x)")
plt.ylabel("x")
plt.grid(True)
plt.legend()
plt.show()

# Plot demonstrating plt.axis('tight')
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='red')
plt.axis('tight')
plt.title("Plot with plt.axis('tight')")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Plot demonstrating plt.axis('equal')
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='purple')
plt.axis('equal')
plt.title("Plot with plt.axis('equal')")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
# 6. Labels and Legends
# Define the functions f(x) = x and g(x) = x^(1/3)
def f(x):
    return x

def g(x):
    return x**(1/3)

# Define the range for x values (x >= 0)
x_values = np.linspace(0, 10, 400)

# Calculate y values for each function
f_values = f(x_values)
g_values = g(x_values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, f_values, label='f(x) = x', color='blue')
plt.plot(x_values, g_values, label='g(x) = x^(1/3)', color='green')

# Add title and labels
plt.title("Plot of f(x) = x and g(x) = x^(1/3)")
plt.xlabel("x")
plt.ylabel("y")

# Add a legend
plt.legend()

# Display the plot with a grid
plt.grid(True)
plt.show()
```

```python
# 7. Basic Scatter Plot

# Generate 50 random points for x and y using np.random.rand()
x_points = np.random.rand(50)
y_points = np.random.rand(50)

# Create scatter plot using plt.plot() with 'o' marker
plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points, 'o', label="plt.plot() with 'o' marker", color='blue')
plt.title("Scatter Plot using plt.plot()")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# Create scatter plot using plt.scatter()
plt.figure(figsize=(8, 6))
plt.scatter(x_points, y_points, label="plt.scatter()", color='green')
plt.title("Scatter Plot using plt.scatter()")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 8. Customizing Scatter Plots

# Define the function f(x) = x * sin(x)
def f(x):
    return x * np.sin(x)

# Generate x values over the range [0, 4]
x_values = np.linspace(0, 4 * np.pi, 100)

# Generate corresponding y values
y_values = f(x_values)

# Create a scatter plot with a line connecting the points
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, linewidth=2, color='blue', label='f(x) = x * sin(x)')
plt.scatter(x_values, y_values,
            s=80,  # Marker size
            facecolor='orange',  # Marker face color
            edgecolor='red',  # Marker edge color
            linewidth=2,  # Marker edge width
            label='Data points')

# Customize the plot
plt.title("Scatter Plot of f(x) = x * sin(x) with Custom Markers and Line")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# 9. Combined Plot

# Define the function f(x) = x * cos(x)
def f(x):
    return x * np.cos(x)

# Generate x values over the range [-2, 2]
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Generate corresponding y values
y_values = f(x_values)

# Define x values at every /4 interval
scatter_x_values = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
scatter_y_values = f(scatter_x_values)

# Create the plot with different styles
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, linestyle='-', color='blue', linewidth=2, label='f(x) = x * cos(x)')
plt.scatter(scatter_x_values, scatter_y_values,
            color='red', marker='o', s=100,  # Custom scatter point style
            edgecolor='black', label='Scatter points at /4 intervals')

# Customize the plot
plt.title("Plot of f(x) = x * cos(x) with Scatter Points at /4 Intervals")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# 10. Custom Plot with Alternative Style

# Define the functions
def g(x):
    return x * np.exp(-x**2)

def h(x):
    return np.sinc(x / np.pi)  # Handling x = 0, sinc function is sin(x)/x

def j(x):
    return 1 / (1 + np.exp(-x))  # Logistic function

# Generate x values for the plot
x_values = np.linspace(-5, 5, 400)

# Calculate y values for each function
g_values = g(x_values)
h_values = h(x_values)
j_values = j(x_values)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot g(x) = x * e^(-x^2) with a custom line style and color
plt.plot(x_values, g_values, linestyle='--', color='blue', linewidth=2, label='g(x) = x * e^(-x^2)')

# Plot h(x) = sin(x)/x (using sinc) with scatter points and a custom marker
plt.plot(x_values, h_values, linestyle='-', color='green', linewidth=2, label='h(x) = sin(x)/x')
plt.scatter(x_values, h_values, color='red', s=40, edgecolor='black', label='Scatter points for h(x)')

# Plot j(x) = 1 / (1 + e^(-x)) with a custom color and line width
plt.plot(x_values, j_values, linestyle='-.', color='purple', linewidth=2, label='j(x) = 1 / (1 + e^(-x))')

# Customize the plot appearance
plt.title("Custom Plot of Multiple Functions", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.xlim(-5, 5)
plt.ylim(-1.5, 1.5)
plt.grid(True)

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# 11. Style Comparison

# Define the function f(x) = x * cos(x)
def f(x):
    return x * np.cos(x)

# Generate x values over the range [-2, 2]
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Generate corresponding y values
y_values = f(x_values)

# Define x values at every /4 interval for scatter points
scatter_x_values = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
scatter_y_values = f(scatter_x_values)

# 1. Plot with the default style
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, linestyle='-', color='blue', linewidth=2, label='f(x) = x * cos(x)')
plt.scatter(scatter_x_values, scatter_y_values, color='red', s=100, label='Scatter points', edgecolor='black')
plt.title("Default Style: f(x) = x * cos(x) with Scatter Points")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# 2. Plot with 'seaborn-whitegrid' style
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, linestyle='-', color='blue', linewidth=2, label='f(x) = x * cos(x)')
plt.scatter(scatter_x_values, scatter_y_values, color='red', s=100, label='Scatter points', edgecolor='black')
plt.title("Seaborn-Whitegrid Style: f(x) = x * cos(x) with Scatter Points")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# 3. Plot with alternative 'ggplot' style
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, linestyle='-', color='blue', linewidth=2, label='f(x) = x * cos(x)')
plt.scatter(scatter_x_values, scatter_y_values, color='red', s=100, label='Scatter points', edgecolor='black')
plt.title("Alternative Style (ggplot): f(x) = x * cos(x) with Scatter Points")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# Reset to default style for further tasks
plt.style.use('default')
```
